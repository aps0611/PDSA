```python
'''
The last non-leaf node is at index (n-2)//2
n = len(A)
'''


def minheap(A):
  for i in range((len(A)-2)// 2  + 1):
    if (A[i] > A[2*i + 1]) or (2*i + 2 != len(A) and A[i] > A[2*i + 2]):
      return False
  return True

def maxheap(A):
  for i in range((len(A)-2)//2 + 1):
    if (A[i] < A[2*i + 1]) or (2*i+2 != len(A) and A[i] < A[2*i + 2]):
      return False
  return True

def typeOfHeap(A):
  if minheap(A) == True:
    return 'Min'
  if maxheap(A) == True:
    return 'Max'
  return 'None'
```

```python
A=eval(input())
print(typeOfHeap(A))
```
