```python
def bellmanFord(WList, s):

  '''Maintain Two dictionaries'''
  distance = {}
  previous = {}

  '''Initially the distance is to infinity'''
  INF = 9999999999999999999

  for v in WList.keys():
    distance[v] = INF
    previous[v] = None

  '''set the initial source vertex to zero distance'''
  distance[s] = 0

  for i in WList.keys():
    for u in WList.keys():
      for (v,d) in WList[u]:
        if distance[v] > distance[u] + d:
          distance[v] = distance[u] + d
          previous[v] = u

  return (distance, previous)


def minCost(route_map, source, destination):
  distance, path = bellmanFord(route_map, source, destination)

  total_distance = distance[destination]
  Route = []

  if distance[destination] != 0:
    dest = destination
    # backtracking
    while dest != source:
      Route = [dest] + Route
      for i, j in path.items():
        if dest == i:
          dest = j
          break

    Route = [dest] + Route

  return (total_distance, Route)
```
