```python
def IsNegativeWeightCyclePresent(WList):
  '''create a dictionary'''
  distance = {}

  '''set initial source index to be s = 0'''
  s = 0

  INF = 99999999999999999999

  for v in WList.keys():
    distance[v] = INF

  distance[s] = 0

  for i in WList.keys():
    for u in WList.keys():
      for v,d in WList[u]:
        distance[v] = min(distance[u] + d,distance[v])

  # After completing the bellman ford. The if condition should be False.
  # if true the negative edge cycle exists!!!
  for u in WList.keys():
    for v,d in WList[u]:
      if distance[v] > distance[u] + d:
        return True

  return False

```
