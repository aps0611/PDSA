```python
def FiberLink(WList):

  '''Initialise three variables- visited, dictionary, Edges'''
  visited = {}
  distance = {}
  edges = []

  INF = 9999999999999999999999

  '''Initially set the visited to False and distance to infinity'''
  for v in WList.keys():
    visited[v] = False
    distance[v] = INF

  '''For the first node 0'''
  visited[0] = True
  for (v,d) in WList[0]:
    distance[v] = d

  '''Visit each node'''
  for i in WList.keys():
    '''set the mindistance = inf and next vertex = None Initially'''
    minDist, nextv = INF, None
    '''visit vertex u'''
    for u in WList.keys():
      '''check neighbors of vertex u i.e. v and distance d resp'''
      for (v,d) in WList[u]:
        '''update as soon as found the minDistance'''
        if visited[u] and (not visited[v]) and d < minDist:
          minDist,nextv, nexte = (d,v,(u,v))

    if nextv is None:
      break

    visited[nextv] = True
    edges.append(nexte)

    for (v,d) in WList[nextv]:
      if not visited[v]:
        distance[v] = min(distance[v], d)

  ans = 0
  for c in edges:
    for v, dist in WList[c[0]]:
      if v == c[1]:
        ans += dist

  return ans
```
