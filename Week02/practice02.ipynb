```python
L = [2,6,8,11,17,23,33,44,46,50,65]
```

```python

def binarySearchIndexAndComparison(L,k):
  n = len(L)

  if n <1:
    return (False, 0)

  ## Philosophy behind binary search
    ## if mid == element its found
    ## else if its less shift the right and left pointers

  # my current location of pointers:
  rPoint = n-1
  lPoint = 0
  count = 0

  while lPoint < rPoint:
    count += 1

    mid = (lPoint + rPoint)//2

    ## case 1:
    if L[mid] == k:
      return (True,count)

    ## case 2:
    elif k < L[mid]:
      rPoint = mid - 1

    ## case 3:
    else:
      lPoint = mid + 1

  return (False, count)

```

```python
binarySearchIndexAndComparison(L,11)
```

```python
def sortInRange(L,r):
  '''
  Create a dictionary with r keys for each integer in range r, initialize
  every value to 0 '''

  countDict = dict.fromkeys(range(r), 0)

  # Iterate over the array and count each integer in the list
  for num in L:
    countDict[num] += 1

  index = 0

  for key in countDict.keys():
    for i in range(countDict[key]):
      L[index] = key
      index += 1

  return L
```

```python
sortInRange([2, 0, 1, 1, 2, 3, 0, 2, 1, 0, 2, 3, 1, 2],4)
```
