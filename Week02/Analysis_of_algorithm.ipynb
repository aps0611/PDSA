# Measuring Performance:

Two main resource of interest:

1. **Running Time**: How long algorithm takes to run | depends on processing power
    * Depends on Input Size- E.g. Larger arrays take longer to sort
    * Time Efficiency = Function(n=Input size) = **t(n)**

2. **Space**: Memory requirement - Availability

Example 1: Sim Card Vs Aadhar Card:

1. Naive Algorithm: $t(n) = n^2$
2. Clever Algorithm: $t(n) = $n * log2(n)$   = number of times to divide n by 2 to reach 1

# How do we evaluate and compare different algorithms formally?

# What is input size



What is natural parameter?

* Size of a list/array that we want to search or sort
* Number of objects we want to rearrange
* Number of vertices and number of edges in the graph

what about the numeric problems? Is n a prime?
* Magnitude of n is not a correct measure
* Arithmetic operations are performed digit by digit
* Number of digits is natural measure of input size

**Summary:**
- Two important parameters when measuring algorithm performance
  - Running time, memory requirement (space)
  - Mainly focus on time
- Running time `t(n)` is a function of input size *n*
  - Interested in orders of magnitude
  - Asymptotic complexity, as *n* becomes large
- From running time, we can estimate feasible input sizes
- We focus on worst case inputs
  - Pessimistic, but easier to calculate than average case
  - Upper bound on worst case gives us an overall guarantee on performance
